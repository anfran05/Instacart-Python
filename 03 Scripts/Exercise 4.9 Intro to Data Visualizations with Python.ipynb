## 4.9 Intro to Data Visualizations with Python

```python
#Import Libraries
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
import scipy
```

```python
#Create path
path = r'/Users/anfran/Documents/Instacart Basket Analysis'
```

```python
#Import data set
ords_prods_merge = pd.read_pickle(os.path.join(path, '02 Data', 'Prepared Data', 'orders_products_combined_updated_4.8.2.pkl'))
```

#### Bar Charts

```python
#Create Bar Chart for 'orders_day_of_week'
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(rot=45, figsize = (10,5))
plt.ylabel("Orders Placed (in millions)")
plt.title ("Instacart Orders by Day")
plt.xticks([0,1,2,3,4,5,6],['Saturday','Sunday','Friday',
            'Monday','Thursday','Tuesday',
            'Wednesday'])
```

```python
#Export bar chart
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow_new.png'))
```

```python
#Create Bar Chart colors for 'orders_day_of_week'
ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color = ['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
#Define 'bar'
bar = ords_prods_merge['orders_day_of_week'].value_counts().plot.bar(color =['purple', 'red', 'pink', 'orange', 'yellow', 'green', 'blue'])
```

```python
bar.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'bar_orders_dow.png'))
```

```python
#Create a list holding True/False values to test np.random.rand() <= 0.7
dev = np.random.rand(len(ords_prods_merge)) <= 0.7
```

```python
#Store 70% of the sample in the dataframe big
big = ords_prods_merge[dev]
```

```python
#Store 30% of the sample in the dataframe small
small = ords_prods_merge[~dev]
```

```python
#Check results
len(ords_prods_merge)
```

```python
len(big) + len(small)
```

```python
#Reduce samples to only those columns necessary for the chart
df_2 = small[['orders_day_of_week','prices']]
```

```python
#Create line chart
line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#Define lnchart
lnchart = line = sns.lineplot(data = df_2, x = 'orders_day_of_week',y = 'prices')
```

```python
#Export line chart
lnchart.figure.savefig(os.path.join(path, '04 Analysis','Visualizations', 'lnchart_prices_dow.png'))
```
